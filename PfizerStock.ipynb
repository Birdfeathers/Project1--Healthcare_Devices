```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

## 2021 Data

```python
pfizer_2021_csv = Path('Pfizer2021.csv')

pfizer_2021_stock_df = pd.read_csv(pfizer_2021_csv)
```

```python
display(pfizer_2021_stock_df)
```

```python
pfizer_2021_stock_df['Date'] = pd.to_datetime(pfizer_2021_stock_df['Date'])
print(pfizer_2021_stock_df.dtypes)
pfizer_2021_stock_df.head()
```

```python
pfizer_2021_stock_df_for_graph = pfizer_2021_stock_df.set_index('Date')

pfizer_2021_stock_df_for_graph.plot()
```

```python
clean_pfizer_2021_stock_df = pfizer_2021_stock_df.reset_index()
clean_pfizer_2021_stock_df['week'] = clean_pfizer_2021_stock_df['Date'].dt.isocalendar().week
clean_pfizer_2021_stock_df['year'] = clean_pfizer_2021_stock_df['Date'].dt.year
print(clean_pfizer_2021_stock_df.head())
```

## 2022 Data

```python
pfizer_2022_csv = Path('Pfizer2022.csv')

pfizer_2022_stock_df = pd.read_csv(pfizer_2022_csv)
```

```python
display(pfizer_2022_stock_df)
```

```python
pfizer_2022_stock_df['Date'] = pd.to_datetime(pfizer_2022_stock_df['Date'])
print(pfizer_2022_stock_df.dtypes)
pfizer_2022_stock_df.head()
```

```python
pfizer_2022_stock_df_for_graph = pfizer_2022_stock_df.set_index('Date')

pfizer_2022_stock_df_for_graph.plot()
```

```python
clean_pfizer_2022_stock_df = pfizer_2022_stock_df.reset_index()
clean_pfizer_2022_stock_df['week'] = clean_pfizer_2022_stock_df['Date'].dt.isocalendar().week
clean_pfizer_2022_stock_df['year'] = clean_pfizer_2022_stock_df['Date'].dt.year
print(clean_pfizer_2021_stock_df.head())
```

## Merge

```python
#new_df = pd.merge(clean_pfizer_2021_stock_df, clean_pfizer_2022_stock_df, on=['week', 'year'])


combined_pfizer_df = pd.concat([pfizer_2021_stock_df, pfizer_2022_stock_df], ignore_index=True)
combined_pfizer_df.head()
#new_df.info()
```

```python
combined_pfizer_df = combined_pfizer_df.set_index('Date')

print(combined_pfizer_df.info)
```

```python
combined_pfizer_df = combined_pfizer_df.dropna()
print(combined_pfizer_df)
```
